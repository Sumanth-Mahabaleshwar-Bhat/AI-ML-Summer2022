```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def mean_Calculation(file):
    data = pd.read_excel (file , header = None)
    data_np = pd.DataFrame.to_numpy(data)
    mean = np.mean(data_np, axis = 0)
    return mean
```

```python
def covariance_Matrix(file):
    data = pd.read_excel (file , header = None)
    data_np = pd.DataFrame.to_numpy(data)
    q = data_np - mean
    shape = data_np.shape
    cov_matrix = np.dot(np.transpose(q),(q))/(shape[0])
    return cov_matrix
```
